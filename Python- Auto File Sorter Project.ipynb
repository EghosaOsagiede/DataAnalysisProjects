# Automatic File Sorter For File Explorer

```python
import os, shutil
```

```python
path = r"C:/Users/DELL/Downloads/New folder for practice/"
```

```python
os.listdir(path)
```

```python
file_names = os.listdir(path)
```

```python
os.path.exists(path + 'csv files')
```

```python
# or
folder_names = ['image files', 'video files', 'csv files']
for loop in range(0,3):
   if not os.path.exists(path + folder_names[loop]):
       os.makedirs(path + folder_names[loop])
       print((path + folder_names[loop]))
```

```python
# OR you can decide to ignore the loop function and do it individually like the one below
```

```python
if not os.path.exists(path + 'image files'):
    os.makedirs(path + 'image files')
    print((path + 'image files'))
if not os.path.exists(path + 'video files'):
    os.makedirs(path + 'video files')
    print(path + 'video files')
if not os.path.exists(path + 'csv files'):
    os.makedirs(path + 'csv files')
    print((path + 'csv files'))
    
```

```python
for file in file_names:
    if ".csv" in file and not os.path.exists(path + "csv files/" + file):
        shutil.move(path + file, path + "csv files/" + file)
        print((file, 'to', path + "csv files/" + file))
    if ".JPG" in file and not os.path.exists(path + "image files/" + file):
        shutil.move(path + file, path + "image files/" + file)
        print((file, 'to', path + "image files/" + file))
    if ".MP4" in file and not os.path.exists(path + "video files/" + file):
        shutil.move(path + file, path + "video files/" + file)
        print((file, 'to', path + "video files/" + file))
```
